```python
import requests
import csv
import re
import itertools
import pickle
```

```python
def alt_element(a):
    return a[::2]

def make_gene_dictionary():
    resp = requests.get('http://www.cbioportal.org/webservice.do?cmd=getTypesOfCancer')
    temp = list(itertools.chain.from_iterable(
        [ x.splitlines() for x in re.split(r'\t+', resp.text.rstrip('\t')) ]))
    abbrv = alt_element(temp)[1:]
    name = [ x.lower() for x in alt_element(temp[1:])[1:] ]
    return dict(zip(name, abbrv))
```

```python
dictionary = make_gene_dictionary()
with open('gene_dict.pickle', 'wb') as handle:
    pickle.dump(dictionary, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
with open('gene_dict.pickle', 'rb') as handle:
    usethisdict = pickle.load(handle)
```
